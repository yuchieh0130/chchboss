## Import Track Data

```python
import pymysql
import MySQLdb
import pandas as pd
import numpy as np
```

```python
conn = pymysql.connect(host='127.0.0.1', user='APP',password='bunnytrack',database='mo')
cur = conn.cursor()
sql = "SELECT * FROM track_test"
cur.execute(sql)
results = cur.fetchall()
conn.commit()

#print(results)

user_id = []
start_date = []
end_time = []
end_date = []
weekday = []
start_time = []
category_id = []
map_id = []
place_id = []


for row in results:
    user_id.append(row[1])
    start_date.append(row[2])
    start_time.append(row[3])
    weekday.append(row[4])
    end_date.append(row[5])
    end_time.append(row[6])
    category_id.append(row[7])
    map_id.append(row[8])
    place_id.append(row[9])

all_data = {
    "user_id" : user_id,
    "start_date" : start_date,
    "start_time" : start_time,
    "weekday" : weekday,
    "end_date" : end_date,
    "end_time" : end_time,
    "category_id" : category_id,
    "map_id" : map_id,
    "place_id" : place_id,
}
df = pd.DataFrame(all_data)
df.head()
```
